```python
#Indexing of a string

f1 = ("hello rohan")

print(f1[0])   # 0 is first letter of string
print(f1[1])
print(f1[0:4]) # 0 to 4 lettr use stop indictor
print(f1[2:8])
print(f1[0:11:2]) # 0 to 11 count and third step
len(f1)   #len function total string count
```

```python
# format() method


txt = "Hello {word}"
print(txt.format(word = 'World!'))
message1 = 'Hi, My name is {} and I am {} years old.'
print(message1.format('rohan', 26))
message2 = 'Hi, My name is {name} and I am {number} years old.'
print(message2.format(name ='rohan', number = 26))
message3 = 'Hi, My name is {0} and I am {1} years old.'
print(message3.format('rohan', 26))
```
