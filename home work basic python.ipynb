.array **atributes**

```python

r5 = np.array([[1, 2, 3],
                [4, 2, 5]])

# Printing type of arr object
print("Array is of type: ", type(r5))

# array dimensions (axes)
print( r5.ndim)

#  shape of array
print( r5.shape)

#  size (total number of elements) of array
print( r5.size)

# type of elements in array
print( r5.dtype)
```

**Array methods**

```python
import numpy as np
array = np.arange(10)
print(array)

r1 = np.mean(array)
print("\nMean: ", r1)

r2 = np.median(array)
print("\nMedian: ", r2)

r3 = np.std(array)
print("\nStandard Deviation: ", r3)
```
